## <span style="background-color: yellow">프로젝트 개요</span>

본 프로젝트는 국내 중소·중견 반도체 기업을 대상으로, 워렌 버핏의 가치투자 철학에 기반한 투자 판단 모델을 구축하는 것을 목표로 한다. 버핏이 강조한 수익성, 안정성, 성장성이라는 세 가지 핵심 기준을 재무지표로 수치화하고, 여기에 브랜드 평판 지수와 경영진 평가 같은 정성적 데이터를 추가하여 종합적인 투자 매력도를 분석하였다.

분석의 흐름은 다음과 같다. 먼저, DART 공시 데이터를 활용해 각 기업의 재무제표를 수집하고, 이를 바탕으로 수익성, 안정성, 성장성 점수를 계산하였다. 이후, 브랜드 이미지와 경영진에 대한 기사 감성 분석 결과를 정성적 지표로 통합하여 최종 데이터셋을 구성하였다.

이 데이터를 기반으로 로그 변환과 표준화를 통해 스케일을 맞추고, KMeans 클러스터링을 적용하여 유사한 특성의 기업군으로 분류하였다. 클러스터 수는 실루엣 점수를 통해 최적화하였으며, PCA를 활용하여 클러스터링 결과를 시각적으로 해석하였다.

마지막으로, 각 클러스터의 평균 지표를 분석해 가치주, 성장주, 저성장군 등의 투자 성향으로 유형화하였으며, 이를 바탕으로 기업별 투자 판단을 제시하였다. 단순한 수익률 중심의 분석을 넘어, 장기적 가치투자 관점에서 신뢰할 수 있는 판단 근거를 제공하고자 하였다.


## <span style="background-color: yellow"> 기존 가치투자 기준에 따른 투자 판단</span>
본 프로젝트는 워런 버핏의 가치투자 철학을 기반으로, 기업의 본질적 가치를 정량적으로 평가하는 방식에서 출발하였다. 이 단계에서는 브랜드 이미지나 경영진 평가와 같은 정성적 요소는 고려하지 않고, 순수하게 재무지표 중심의 수치 기반 분석을 통해 투자 판단 기준을 설정하였다.

버핏의 투자 철학에 따라, 다음의 세 가지 축을 중심으로 기업의 투자 매력도를 평가하였다.

1. 수익성
ROE(자기자본이익률): 자기자본 대비 순이익의 비율로, 기업이 주주가치를 얼마나 창출하고 있는지를 평가

ROA(총자산이익률): 자산을 얼마나 효율적으로 활용하고 있는지를 나타내는 지표

영업이익률: 기업의 본업이 얼마나 안정적으로 이익을 내고 있는지를 반영

→ 높은 수익성은 기업의 지속적인 경쟁력과 수익 창출 능력을 의미한다.

2. 안정성
유동비율: 유동자산 대비 유동부채의 비율로, 단기 채무 상환 능력을 의미

부채비율: 전체 자본 구조에서 부채가 차지하는 비중으로, 재무 건전성을 판단하는 핵심 지표

→ 안정적인 재무구조는 리스크를 줄이고 장기 투자에 대한 신뢰도를 높인다.

3. 성장성
매출액 성장률: 시장 확대 가능성과 외형 성장성 평가

순이익 성장률: 기업의 수익성이 얼마나 개선되고 있는지를 반영

→ 지속적인 성장세는 장기 투자 시 복리 효과를 기대할 수 있는 중요한 조건이다.

이러한 수익성, 안정성, 성장성 지표는 워런 버핏이 실제로 기업을 분석할 때 중시하는 핵심 기준이며, 본 프로젝트 초기 분석 모델은 이 세 가지 지표만을 바탕으로 기업의 투자 매력도를 클러스터링하고 해석하였다.

※ 이후 단계에서는 이 정량적 평가만으로는 포착하기 어려운 기업의 브랜드 가치나 경영진의 질적 역량까지 고려하기 위해 정성적 요소를 모델에 추가하였다.

## <span style="background-color: yellow"> 기존 모델과의 차이점: 정성적 지표의 반영</span>
기존의 워런 버핏 가치투자 기준은 ROE, 부채비율, 성장률 등 수치 기반의 정량적 지표를 중심으로 기업의 내재 가치를 평가하는 방식이었다. 하지만 정량 지표만으로는 기업의 비재무적 요소나 외부 리스크 요인을 완전히 반영하기 어렵다는 한계가 존재한다.

이를 보완하기 위해 본 프로젝트에서는 다음과 같은 정성적 요소를 추가하였다.

1. 브랜드 평판 지수
소비자 인식, 언론 노출, 사회적 이미지 등을 종합하여 기업의 대외적 신뢰도를 평가하는 지표

기업의 시장 신뢰도와 지속 가능성을 나타내는 대표적인 정성적 지표

워런 버핏이 강조한 "경제적 해자(Moat)" 개념과 맞닿아 있으며, 장기 경쟁우위를 보여주는 핵심 요소로 간주된다

2. 경영진 평가
기업 관련 뉴스 기사에 등장하는 경영진 키워드(예: 리더십, 윤리성, 전략 등)에 대한 감성 분석을 실시

CEO 및 주요 경영진의 이미지, 신뢰도, 장기 전략 역량 등을 정성적으로 판단

워런 버핏 역시 기업 분석 시 “경영진의 정직함과 유능함”을 중시한다고 강조

왜 정성적 기준이 중요한가?
고성장 기업일수록 외부 변수(경쟁, 규제, 사회적 이슈)에 민감하며,
단순한 재무 지표만으로는 잠재 리스크나 기회 요인을 충분히 파악하기 어렵다.

특히 브랜드 가치나 경영진의 신뢰도는 수치화되기 어렵지만,
기업의 지속 성장 가능성과 위기 대응 능력에 결정적 영향을 미칠 수 있다.

따라서 본 프로젝트는 단순한 정량 모델에 그치지 않고, 브랜드 평판 지수와 경영진 평가를 통합하여
보다 현실적이고 신뢰도 높은 가치투자 모델로 확장하였다.

````python

# 필수 라이브러리 임포트
import requests  # DART API에 요청 보내는 용도 (웹에서 데이터 가져오기)
import pandas as pd  # 테이블(데이터프레임) 형태로 데이터 처리
import time  # 요청 간 시간 지연(딜레이) 주기 위해 사용

#  DART API KEY
API_KEY = '24efd070196004e8f8e8c754e59717bf475f5e43'


# 기업명과 DART에서 사용하는 고유 기업 코드 매핑 (기업명 → corp_code)
# 우리가 분석할 반도체 관련 기업들의 코드 모아놓음
corp_code_dict = {
     '삼성전자': '00126380',
    'SK하이닉스': '00164779',
    'DB하이텍': '00160843',
    '하나마이크론': '00445054',
    '성우테크론': '00350482',
    '이오테크닉스': '00246417',
    '에이디테크놀로지': '00657002',
    '어보브반도체': '00591441',
    'LX세미콘': '00525934',
    '텔레칩스': '00423177',
    '주성엔지니어링': '00252135',
    '지니틱스': '01318933',
    '두산테스나': '00563545',
    'ISC': '00572905',
    '젬백스': '00492894',
    '제우스': '00442570',
    '제주반도체': '00447487',
    '와이씨': '01109539',
    '동진쎄미켐': '00118804',
    '솔브레인': '01489648',
    '유진테크': '00531014',
    '디아이': '00118275',
    '네온테크': '01325429',
    '미코': '00366942',
    '네오셈': '01170865',
    '신성이엔지': '00136271',
    '칩스앤미디어': '00579971',
    '해성디에스': '01032413',
    'SFA반도체': '00301246',
    '에이팩트': '00667425',
    '하나머티리얼즈': '00660750',
    '에스앤에스텍': '00411048',
    '코미코': '00997812',
    'GST': '00540863',
    '원익QnC': '00468374',
    '티씨케이': '00245472',
    'KEC': '00589127',
    '넥스틴': '01080252',
    '네패스': '00227333',
    '에프에스티': '00223434',
    '엠케이전자': '00121686',
    '유니퀘스트': '00414601',
    '윈팩': '00536286',
    '예스티': '00530556',
    '유니테스트': '00540447',
    '엑시콘': '00611736',
    '큐에스아이': '00411446',
    '유니셈': '00260569',
    '덕산하이메탈': '00471068',
    '오로스테크놀로지': '01016479',
    '인텍플러스': '00479787',
    '티에스이': '00372226',
    '미래산업': '00197759',
    '월덱스': '00580065',
    '시그네틱스': '00158219',
    '엘오티베큠': '00484682',
    '유니트론텍': '00879020',
    '네패스아크': '01368354',
    '티엘비': '00920379',
    '케이씨': '00209780',
    '아진엑스텍': '00404288',
    '오킨스전자': '00543107',
    '코세스': '00365934',
    'LB세미콘': '00412348',
    '뉴파워프라즈마': '00521390',
    '한솔아이원스': '00625942',
    '한양디지텍': '00488989',
    '씨앤지하이테크': '00605328',
    '디엔에프': '00569691',
    '메카로': '00652007',
    '엔투텍': '01105621',
    '아이텍': '00626011',
    '싸이맥스': '00659161',
    '서플러스글로벌': '00599285',
    '테스': '00524421',
    '원익IPS': '01135941',
    '한미반도체': '00161383',
    '피델릭스': '00189538',
    '픽셀플러스': '00495086',
    '제너셈': '00666204',
    '타이거일렉': '00617314',
    '오디텍': '00529815',
    'KX하이텍': '00276834',
    '엑사이엔씨': '00360577',
    '티이엠씨씨엔에스': '01113754',
    '엔시트론': '00641393'
}

# 분석할 연도 설정 (2021~2025년)
years = [str(y) for y in range(2021, 2026)]

# 분기 코드 설정 (DART API가 요구하는 코드 형식으로 설정)
quarter_map = {
    'Q1': '11013',
    'Q2': '11012',
    'Q3': '11014',
    'Q4': '11011'
}

# 결과 저장용 리스트 (데이터 누적 저장)
records = []

# 주식 수 가져오는 보조 함수 (기본적으로 사용하는 함수)
def get_number_of_shares(corp_code, year, reprt_code):
    try:
        url = 'https://opendart.fss.or.kr/api/stockTotqySttus.json'
        params = {
            'crtfc_key': API_KEY, 'corp_code': corp_code,
            'bsns_year': year, 'reprt_code': reprt_code
        }
        res = requests.get(url, params=params)
        data = res.json()
        for item in data.get('list', []):
            # '보통주' 수만 가져오기
            if '보통주' in item['se'] and item['istc_totqy'] != '-':
                return int(item['istc_totqy'].replace(',', ''))
    except:
        return None
    return None

# Fallback용 주식 수 가져오는 함수
# 2025년 Q1 데이터가 부족할 경우 2024년 Q4 데이터 사용
def get_number_of_shares_fallback(corp_code, year, reprt_code):
    try:
        url = 'https://opendart.fss.or.kr/api/stockTotqySttus.json'
        params = {
            'crtfc_key': API_KEY, 'corp_code': corp_code,
            'bsns_year': year, 'reprt_code': reprt_code
        }
        res = requests.get(url, params=params)
        data = res.json()
        for item in data.get('list', []):
            if '보통주' in item['se'] and item['istc_totqy'] != '-':                
                return int(item['istc_totqy'].replace(',', ''))
    except Exception as e:
        print(f'주식수 fallback 오류: {e}')
        return None
    return None

# 재무제표 가져오는 메인 함수
# 핵심 재무지표(매출액, 순이익 등)를 DART에서 가져와서 계산
def get_financials(corp_name, corp_code, year, quarter_name, reprt_code):
    try:
        url = 'https://opendart.fss.or.kr/api/fnlttSinglAcnt.json'
        params = {
            'crtfc_key': API_KEY, 'corp_code': corp_code,
            'bsns_year': year, 'reprt_code': reprt_code,
            'fs_div': 'CFS'  # 연결재무제표 사용
        }
        res = requests.get(url, params=params)
        data = res.json()

        if data.get('status') != '000': return None  # 데이터 없으면 종료

        # 원하는 재무 항목만 뽑아서 저장
        accounts = {
            '매출액': 0, '영업이익': 0, '당기순이익': 0,
            '자산총계': 0, '부채총계': 0, '자본총계': 0,
            '유동자산': 0, '유동부채': 0
        }

        # 재무제표 항목 파싱해서 accounts dict에 채우기
        for item in data.get('list', []):
            name = item['account_nm']
            amt = item['thstrm_amount'].replace(',', '').replace('-', '0')
            amt = int(amt) if amt.isdigit() else 0
            for key in accounts:
                if key in name and accounts[key] == 0:
                    accounts[key] = amt

        # 주식 수 가져오기 (2025년 Q1 예외 처리 포함)
        if year == '2025' and reprt_code == '11013':
            shares = get_number_of_shares_fallback(corp_code, '2024', '11011')
        else:
            shares = get_number_of_shares(corp_code, year, reprt_code)

        # 계산 결과 구성
        result = {
            '기업명': corp_name,
            '분기라벨': f"{year}_{quarter_name}",
            'ROE': round(accounts['당기순이익'] / accounts['자본총계'] * 100, 2) if accounts['자본총계'] else None,
            'ROA': round(accounts['당기순이익'] / accounts['자산총계'] * 100, 2) if accounts['자산총계'] else None,
            '영업이익률': round(accounts['영업이익'] / accounts['매출액'] * 100, 2) if accounts['매출액'] else None,
            '순이익률': round(accounts['당기순이익'] / accounts['매출액'] * 100, 2) if accounts['매출액'] else None,
            'EPS': round(accounts['당기순이익'] / shares, 2) if shares else 'N/A',
            '부채비율': round(accounts['부채총계'] / accounts['자본총계'] * 100, 2) if accounts['자본총계'] else None,
            '유동비율': round(accounts['유동자산'] / accounts['유동부채'] * 100, 2) if accounts['유동부채'] else None,
            '유동자산': accounts['유동자산'],
            '자기자본비율': round(accounts['자본총계'] / accounts['자산총계'] * 100, 2) if accounts['자산총계'] else None,
            '매출액': accounts['매출액'],
            '당기순이익': accounts['당기순이익']
        }
        return result

    except Exception as e:
        print(f'get_financials 오류: {e}')
        return None

# 메인 수집 루프
# 기업별 / 연도별 / 분기별로 반복하면서 데이터 수집
for corp_name, corp_code in corp_code_dict.items():
    for year in years:
        for qtr_name, reprt_code in quarter_map.items():
            # 2025년은 Q1까지만 수집
            if year == '2025' and qtr_name != 'Q1':
                continue            
            row = get_financials(corp_name, corp_code, year, qtr_name, reprt_code)
            if row:
                records.append(row)
            else:
                print(f" ")
            time.sleep(1.0)  # 서버 과부하 방지용 1초 딜레이

# 수집한 데이터를 DataFrame으로 변환
df = pd.DataFrame(records)

# EPS 전처리
# 숫자로 변환 후, 연도별 평균으로 결측치 보완
df['EPS'] = pd.to_numeric(df['EPS'], errors='coerce')
df['연도'] = df['분기라벨'].str.split('_').str[0]
df['EPS'] = df.groupby(['기업명', '연도'])['EPS'].transform(lambda x: x.fillna(x.mean()))
df = df.drop(columns=['연도'])

# 성장성 계산
# 매출액 성장률, 순이익 성장률 계산 (% 단위)
df['매출액 성장률'] = df.groupby('기업명')['매출액'].pct_change() * 100
df['순이익 성장률'] = df.groupby('기업명')['당기순이익'].pct_change() * 100

# 분석 대상 분기만 남기기 (2022년부터 시작)
df = df[df['분기라벨'].str.startswith('2022') |
        df['분기라벨'].str.startswith('2023') |
        df['분기라벨'].str.startswith('2024') |
        df['분기라벨'].str.startswith('2025')]

# 최종 DataFrame 컬럼 순서 정렬
df = df[['기업명', '분기라벨', 'ROE', 'ROA', '영업이익률', '순이익률', 'EPS',
         '부채비율', '유동비율', '유동자산', '자기자본비율',
         '매출액', '매출액 성장률', '당기순이익', '순이익 성장률']]
````

````python
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 설정
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False 
````

## 한글 폰트 설정

````python
import warnings
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak on Windows with MKL*")


# 버핏식 점수 계산
# → 수익성, 안정성, 성장성 점수를 각각 계산해서 기업별 종합 투자 판단에 활용

# 수익성점수 = ROE, ROA, 영업이익률, 순이익률, EPS 의 평균
df['수익성점수'] = df[['ROE', 'ROA', '영업이익률', '순이익률', 'EPS']].apply(lambda x: x.mean(), axis=1)

# 안정성점수 = 유동비율과 자기자본비율 평균에서 부채비율을 빼서 계산 (부채 많으면 점수 낮아짐)
df['안정성점수'] = df[['유동비율', '자기자본비율']].apply(lambda x: x.mean(), axis=1) - df['부채비율']

# 성장성점수 = 매출액 성장률과 순이익 성장률의 평균
df['성장성점수'] = df[['매출액 성장률', '순이익 성장률']].apply(lambda x: x.mean(), axis=1)

# NaN 제거
# → 점수 계산에 결측치(NaN)가 있는 기업은 제외
score_features = ['수익성점수', '안정성점수', '성장성점수']
df_no_nan = df.dropna(subset=score_features)

# score_features log1p 적용 (극단값 완화)
# → log1p(로그 변환)를 통해 극단적으로 높은 값의 영향을 줄임 (스케일 맞추기용)
import numpy as np
for feature in score_features:
    df_no_nan[feature + '_log'] = np.log1p(df_no_nan[feature].clip(lower=0))  # 음수는 0으로 클리핑

log_features = [f + '_log' for f in score_features]

# scaling 적용 (log_features 3개)
# → StandardScaler로 평균 0, 표준편차 1로 스케일링 (거리 기반 알고리즘에 필요)
from sklearn.preprocessing import StandardScaler
X_score_scaled = StandardScaler().fit_transform(df_no_nan[log_features])

#  silhouette score 탐색
# → 클러스터 개수(k)를 결정하기 위해 silhouette score 를 사용해서 최적 k 찾기
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

scores = []
K_range = range(2, 8)  # k=2 ~ 7 까지 확인

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_score_scaled)
    score = silhouette_score(X_score_scaled, labels)
    scores.append(score)

# 시각화 → k값에 따른 silhouette score 그래프 출력
plt.figure(figsize=(6,4))
plt.plot(K_range, scores, marker='o')
plt.xlabel('클러스터 수 (k)')
plt.ylabel('Silhouette Score')
plt.title('적절한 클러스터 수 탐색 (log1p score_features 기반)')
plt.show()

# 최적 k 선택 (예시로 k=3 사용 → 결과 보고 수정 가능)
# → k=3 으로 클러스터링 적용
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_no_nan['cluster_score_total'] = kmeans.fit_predict(X_score_scaled)

# PCA 2D 적용
# → PCA로 고차원 데이터를 2D로 축소 → 시각화를 위해 사용
from sklearn.decomposition import PCA
from matplotlib.patches import Ellipse

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_score_scaled)

df_no_nan['PCA1'] = X_pca[:,0]
df_no_nan['PCA2'] = X_pca[:,1]

# PCA 기반 scatter + cluster별 원 표시
# → PCA 시각화 (scatter plot) + 클러스터별 원(ellipse) 표시

plt.figure(figsize=(12,10))

# 클러스터별 색 지정
colors = {0: 'red', 1: 'blue', 2: 'green'}

# 클러스터별로 scatter plot 그리기
for cluster in sorted(df_no_nan['cluster_score_total'].unique()):
    subset = df_no_nan[df_no_nan['cluster_score_total'] == cluster]
    plt.scatter(subset['PCA1'], subset['PCA2'],
                color=colors[cluster],
                label=f'Cluster {cluster+1}',
                alpha=0.6,
                s=50)
    
    # 원(ellipse) 표시 (데이터가 2개 이상일 때만)
    if len(subset) >= 2:
        cov = np.cov(subset['PCA1'], subset['PCA2'])
        lambda_, v = np.linalg.eig(cov)
        lambda_ = np.sqrt(lambda_)
        ell = Ellipse(xy=(np.mean(subset['PCA1']), np.mean(subset['PCA2'])),
                      width=lambda_[0]*4, height=lambda_[1]*4,
                      angle=np.rad2deg(np.arccos(v[0, 0])),
                      color=colors[cluster], alpha=0.2)
        plt.gca().add_patch(ell)
    
    # 클러스터 중심에 label 표시
    plt.text(np.mean(subset['PCA1']), np.mean(subset['PCA2']),
             f'Cluster {cluster+1}',
             fontsize=12, weight='bold', ha='center', va='center')

plt.title('기업 클러스터링 결과 (PCA 2D 기반, log1p score_features)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.grid(True)
plt.show()

# 클러스터별 평균 score_features 확인
# → 각 클러스터별로 수익성/안정성/성장성 평균값 확인 (클러스터 특성 파악용)
print("\n클러스터별 평균 수익성/안정성/성장성:")
cluster_summary = df_no_nan.groupby('cluster_score_total')[score_features].mean()
display(cluster_summary)

# 클러스터별 기업 분포 확인
# → 클러스터별로 어떤 기업들이 몇 번씩 나왔는지 확인 (기업별 클러스터 분포 확인용)
print("\n클러스터별 기업 분포:")
cluster_counts = df_no_nan.groupby(['cluster_score_total', '기업명']).size().unstack(fill_value=0)
display(cluster_counts)
````

본 분석에서는 DART API에서 수집한 반도체 기업들의 재무지표 데이터를 활용하여, 기업별로 수익성 / 안정성 / 성장성 점수를 계산하고, 이를 기반으로 유사한 기업군으로 클러스터링을 진행하였다.

일반적으로 투자자들이 참고하는 재무제표 투자지표를 활용하여 기업을 유형화함으로써, 이후 가치투자 판단 모델 개발의 기초 데이터를 구축하였다.

분석 흐름은 이와 같다.

재무지표 수집 → 버핏식 점수 계산 → 로그 변환 → 표준화 → K-Means 클러스터링 → Silhouette Score 평가 → PCA 시각화


<span style="background-color: #00FF00">1. Feature Engineering (버핏식 점수 계산)</span>

 <span style="background-color: yellow">수익성 점수</span>: ROE, ROA, 영업이익률, 순이익률, EPS 의 평균

<span style="background-color: yellow">안정성 점수</span>: 유동비율과 자기자본비율의 평균에서 부채비율 차감

<span style="background-color: yellow">성장성 점수</span>: 매출액 성장률, 순이익 성장률의 평균

활용 이유:
버핏의 가치투자 기준 3가지(수익성, 안정성, 성장성)를 수치화하여 기업별 비교가 용이하도록 구성.

<span style="background-color: #00FF00">로그 변환 (Log1p)</span>

<span style="background-color: yellow"> df_no_nan[feature + '_log'] = np.log1p(df_no_nan[feature].clip(lower=0))</span>

활용 이유:
재무지표는 일반적으로 값의 스케일 차이가 크고, 극단적인 값(Outlier)이 존재할 수 있음 → log1p(자연로그+1) 변환을 통해 분포를 안정화하고 이상치의 영향을 줄임.

<span style="background-color: #00FF00">표준화 (Standard Scaler)</span>
<span style="background-color: yellow">X_score_scaled = StandardScaler().fit_transform(df_no_nan[log_features]) </span>
활용 이유:
클러스터링 알고리즘(K-Means)은 유클리드 거리 기반 → 각 feature의 스케일 차이를 제거하여 공정한 거리 계산 가능하게 하기 위함.

<span style="background-color: #00FF00">K-Means 클러스터링</span> 
<span style="background-color: yellow">kmeans = KMeans(n_clusters=k, random_state=42)</span>
<span style="background-color: yellow">labels = kmeans.fit_predict(X_score_scaled)</span>
활용 이유:
기업들을 "유사한 투자 특성"에 따라 그룹화(클러스터링)하여 유형별 투자 전략 수립 가능.

<span style="background-color: #00FF00">Silhouette Score 평가</span>
<span style="background-color: yellow">silhouette_score(X_score_scaled, labels)</span>
활용 이유:
클러스터링 성능 평가 지표로 사용 → 클러스터가 얼마나 잘 구분되었는지(내부 응집도 vs 클러스터 간 분리도)를 측정.

<span style="background-color: #00FF00">PCA (Principal Component Analysis)</span>
<span style="background-color: yellow">pca = PCA(n_components=2)</span>
<span style="background-color: yellow">X_pca = pca.fit_transform(X_score_scaled)</span>
활용 이유:
고차원 데이터를 2차원으로 축소하여 시각화 → 클러스터링 결과를 직관적으로 확인 가능.

클러스터별 평균 점수 분석: 수익성/안정성/성장성 평균값을 통해 각 클러스터의 특성 파악
클러스터별 기업 분포 확인: 각 클러스터에 속한 기업 리스트 확인 → 특정 투자 성향의 기업군을 식별

일반 투자지표와 가치투자 관점을 통합한 점수 기반 클러스터링을 통해 기업 유형별 분석 기반을 마련함.

````python
import pickle

#  Scaler 저장 (StandardScaler)
# → 학습된 StandardScaler를 'scaler.pkl' 파일로 저장 → 추후 모델 재사용 가능
scaler = StandardScaler()
scaler.fit(df_no_nan[log_features])
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# KMeans 모델 저장
# → 학습된 KMeans 클러스터링 모델을 'kmeans_model.pkl' 파일로 저장 → 추후 모델 재사용 가능
with open('kmeans_model.pkl', 'wb') as f:
    pickle.dump(kmeans, f)

# 클러스터 해석 함수 정의 (워렌 버핏 투자 기준 기반 해석)
# → 클러스터별 평균 점수를 기반으로 해석 문구 자동 생성

def interpret_cluster(cluster_id, cluster_summary):
    # 클러스터별 평균 점수 가져오기
    growth = cluster_summary.loc[cluster_id, '성장성점수']
    profitability = cluster_summary.loc[cluster_id, '수익성점수']
    stability = cluster_summary.loc[cluster_id, '안정성점수']
    
    # 해석 조건 개선

    # 1️⃣ 초고성장 + 초고수익
    if growth > 300 and profitability > 200:
        return "초고성장 초고수익 기업 → 매우 공격적 성장주 (매우 신중 필요, 버핏은 회피)"

    # 2️⃣ 고성장 + 고수익 → 안정성 나쁠 때 별도 처리
    elif growth > 100 and profitability > 100:
        if stability < 0:
            return "고성장 고수익이나 재무 건전성 낮음 → 매우 신중 필요 (버핏은 회피)"
        elif stability < 150:
            return "고성장 고수익 기업 → 신중한 접근 필요 (버핏은 보통 회피)"
        else:
            return "고성장 고수익 + 재무 건전성 양호 → 성장 가치주 (버핏 선호 아님)"

    # 3️⃣ 고성장 + 저수익
    elif growth > 100 and profitability < 100:
        return "고성장 기대주 → 수익성 보완 필요, 리스크 있음 (버핏은 회피)"

    # 4️⃣ 방어적 가치주 → 안정성 높고 수익성도 높음
    elif stability > 150 and profitability > 100:
        return "방어적 가치주 → 워렌 버핏이 선호할 만한 건강한 기업"

    # 5️⃣ 성장성 음수 + 수익성 낮음 → 투자 비추천
    elif growth < 0 and profitability < 100:
        return "투자 비추천 → 성장성, 수익성 모두 낮음"

    # 6️⃣ 그 외 → 투자 추천 또는 중립
    else:
        return "투자 추천 또는 중립 → 추가 분석 필요"



# 삼성전자 row 가져오기
기업_row = df_no_nan[df_no_nan['기업명'] == '주성엔지니어링']

# StandardScaler로 변환
기업_scaled = scaler.transform(기업_row[log_features])

# KMeans 클러스터 예측
기업_cluster = kmeans.predict(기업_scaled)[0]

# 해석 결과 출력
print(f"\n 주성엔지니어링은 클러스터 {기업_cluster} 에 속하며, {interpret_cluster(기업_cluster, cluster_summary)} 입니다.")

# 전체 클러스터 해석 결과 table 출력
# → 전체 클러스터별 평균 점수 + 해석 결과 DataFrame으로 출력

# 결과 저장용 리스트
cluster_interpretations = []

# 클러스터별로 interpret_cluster 호출해서 결과 저장
for cluster_id in sorted(cluster_summary.index):
    interpretation = interpret_cluster(cluster_id, cluster_summary)
    cluster_interpretations.append({
        'Cluster': cluster_id,
        '수익성점수': cluster_summary.loc[cluster_id, '수익성점수'],
        '안정성점수': cluster_summary.loc[cluster_id, '안정성점수'],
        '성장성점수': cluster_summary.loc[cluster_id, '성장성점수'],
        '해석': interpretation
    })

# DataFrame으로 정리
df_cluster_interpretation = pd.DataFrame(cluster_interpretations)

# 결과 출력
print("\n전체 Cluster 해석 결과:")
display(df_cluster_interpretation)
````

따라서 다음 두 가지 정성적 지표를 보완적으로 활용합니다:

브랜드 평판 지수: 시장의 신뢰도 및 기업 이미지

경영진 평가 (기사 기반 감성분석): 리더십 역량과 전략 실행력

이를 통해 고성장 기업의 ‘지속 가능성’과 ‘본질적 경쟁력’을 추가적으로 판단할 수 있습니다.

브랜드 평판 및 경영진 평가 등 정성적 요소 추가

특히 고성장 기업군에 대한 리스크 판단 및 지속 가능성 평가 강화

최종적으로는 정량 + 정성 지표를 통합한
가치투자 판단 프레임워크를 구축하는 것이 목표입니다.

### <span style="background-color: yellow"> 여기서 부터 모델</span>

# 아래 코드를 실행시키고 기업명을 꼭 입력해야 실행됩니다. (ex: 삼성전자, 주성엔지니어링 등)

````python
# 필요한 라이브러리
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 버핏식 클러스터 해석 함수
def interpret_cluster(cluster_id, cluster_summary):
    growth = cluster_summary.loc[cluster_id, '성장성점수']
    profitability = cluster_summary.loc[cluster_id, '수익성점수']
    stability = cluster_summary.loc[cluster_id, '안정성점수']
    
    if growth > 100 and profitability > 100:
        return "공격적 성장주 → 신중한 접근 필요 (버핏은 선호하지 않음)"
    elif growth > 100 and profitability < 100:
        return "고성장 기대주 → 리스크 있음, 신중한 접근 필요 (버핏은 선호하지 않음)"
    elif stability > 150 and profitability > 100:
        return "방어적 가치주 → 워런 버핏의 1차적으로 숫자로 건강한 기업 선정"
    elif growth < 0 and profitability < 100:
        return "투자 비추천"
    else:
        return "투자 추천 또는 중립"

# 기업 분석 함수
def run_investment_model_all_quarters_df_no_nan(company_name):
    기업_row = df_no_nan[df_no_nan['기업명'] == company_name].copy()
    if 기업_row.empty:
        print(f"[오류] 입력한 기업명 '{company_name}'에 해당하는 데이터가 없습니다.")
        return
    
    print(f"\n{company_name} → 총 {len(기업_row)}개 분기 데이터 확인됨.")
    score_features = ['수익성점수', '안정성점수', '성장성점수']
    log_features = [f + '_log' for f in score_features]

    for feature in score_features:
        기업_row[feature + '_log'] = np.log1p(기업_row[feature].clip(lower=0))

    with open('scaler.pkl', 'rb') as f:
        scaler = pickle.load(f)
    with open('kmeans_model.pkl', 'rb') as f:
        kmeans = pickle.load(f)

    X_scaled = scaler.transform(기업_row[log_features])
    기업_row['cluster_score_total'] = kmeans.predict(X_scaled)

    cluster_summary = df_no_nan.groupby('cluster_score_total')[score_features].mean()

    print("\n분기별 클러스터링 결과:")
    for i, row in 기업_row.iterrows():
        label = row['분기라벨']
        cluster = row['cluster_score_total']
        interpretation = interpret_cluster(cluster, cluster_summary)
        print(f"{label}: 클러스터 {cluster} → {interpretation}")

    final_cluster = 기업_row['cluster_score_total'].mode()[0]
    final_interpretation = interpret_cluster(final_cluster, cluster_summary)
    print(f"\n최종 클러스터: {final_cluster} → {final_interpretation}")

    # 클러스터 해석 요약 테이블
    cluster_interpretations = []
    for cluster_id in sorted(cluster_summary.index):
        interpretation = interpret_cluster(cluster_id, cluster_summary)
        cluster_interpretations.append({
            'Cluster': cluster_id,
            '수익성점수': cluster_summary.loc[cluster_id, '수익성점수'],
            '안정성점수': cluster_summary.loc[cluster_id, '안정성점수'],
            '성장성점수': cluster_summary.loc[cluster_id, '성장성점수'],
            '해석': interpretation
        })
    df_cluster_interpretation = pd.DataFrame(cluster_interpretations)
    print("\n전체 Cluster 해석 결과:")
    display(df_cluster_interpretation)

    return 기업_row

# 실행
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
with open('kmeans_model.pkl', 'rb') as f:
    kmeans = pickle.load(f)

company_name = input("기업명을 입력하세요 --> ")
df_result = run_investment_model_all_quarters_df_no_nan(company_name)
````

## 기업별 가치투자 분석 함수 설명
본 코드는 특정 기업을 입력하면, 해당 기업의 모든 분기 데이터를 기반으로
워렌 버핏의 3대 가치 기준(수익성, 안정성, 성장성)에 따라 클러스터링하고,
클러스터별 투자 해석을 제공하는 분석 함수입니다.

1. 함수 개요
run_investment_model_all_quarters_df_no_nan(company_name) 함수는 다음과 같은 과정을 수행합니다.

입력한 기업명에 대해 df_no_nan에서 분기별 데이터를 추출

수익성, 안정성, 성장성 점수를 log1p()로 변환하여 스케일 조정

저장된 StandardScaler 및 KMeans 모델을 불러와 예측 수행

각 분기마다 클러스터 번호를 예측하고, 이에 해당하는 투자 해석을 출력

전체 분기 중 가장 자주 등장한 클러스터를 최종 클러스터로 판단하고, 종합 해석을 제공

2. 클러스터 해석 기준
interpret_cluster() 함수는 클러스터별 평균 점수를 기반으로 다음과 같은 해석을 수행합니다.

성장성↑, 수익성↑: 공격적 성장주 (리스크 있음, 신중한 접근 필요)

성장성↑, 수익성↓: 고성장 기대주 (버핏은 회피)

안정성↑, 수익성↑: 방어적 가치주 (버핏 선호)

성장성↓, 수익성↓: 투자 비추천

그 외: 중립 또는 추가 분석 필요

3. 출력 결과
분기별 클러스터링 결과
각 분기마다 클러스터 번호와 해당 분기에 대한 해석 문구를 출력

최종 클러스터 해석
전체 분기 중 가장 빈도수가 높은 클러스터를 기반으로 해당 기업의 종합적 투자 성향을 해석

클러스터 해석 요약 테이블
모든 클러스터에 대한 평균 점수와 해석 문구를 테이블 형태로 출력하여 비교 가능

이 분석 함수는 특정 기업의 분기별 클러스터 변화와 종합 투자 판단을 함께 확인할 수 있어
기업의 가치 평가 흐름을 시간 단위로 추적하고 비교하는 데 유용합니다.

````python
# Silhouette Plot 시각화 함수 분리
def draw_silhouette_plot(df_no_nan, scaler, kmeans, score_features):
    log_features = [f + '_log' for f in score_features]
    for feature in score_features:
        df_no_nan[feature + '_log'] = np.log1p(df_no_nan[feature].clip(lower=0))

    X_scaled = scaler.transform(df_no_nan[log_features])
    labels = kmeans.predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, labels)
    sample_silhouette_values = silhouette_samples(X_scaled, labels)

    # Plot
    fig, ax1 = plt.subplots(1, 1)
    fig.set_size_inches(8, 6)
    y_lower = 10
    for i in range(kmeans.n_clusters):
        ith_values = sample_silhouette_values[labels == i]
        ith_values.sort()
        size_i = ith_values.shape[0]
        y_upper = y_lower + size_i
        color = cm.nipy_spectral(float(i) / kmeans.n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_values,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_i, str(i))
        y_lower = y_upper + 10

    ax1.set_title("Silhouette Plot")
    ax1.set_xlabel("Silhouette Coefficient Values")
    ax1.set_ylabel("Cluster Label")
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks(np.arange(-0.1, 1.1, 0.2))
    plt.show()

# 실행 예시
score_features = ['수익성점수', '안정성점수', '성장성점수']
draw_silhouette_plot(df_no_nan, scaler, kmeans, score_features)
````

편집

## 4) 사용할 함수들 정의

````python
# 감성분석 점수가 0.6 미만이면 중립, 이상이면 긍정 or 부정
def interpret(label, score, threshold=0.6):
    if score < threshold:
        return "중립"
    return "긍정" if label.lower() == "positive" else "부정"

# df_sum, df_count에 감성점수 누적 및 개수 업뎃
def inData(company: str, year: int, score: float):
    comp = company.replace(" ", "")
    # row 생성
    if comp not in df_sum.index:
        df_sum.loc[comp]   = [np.nan]*len(years)
        df_count.loc[comp] = [0]*len(years)
    # 값 누적 & 개수
    if pd.isna(df_sum.at[comp, year]):
        df_sum.at[comp, year]   = score
        df_count.at[comp, year] = 1
    else:
        df_sum.at[comp, year]   += score
        df_count.at[comp, year] += 1
````

````python
from docx import Document

# docs 파일을 company, year, article 단위로 파싱하여 감성분석 후 데이터 프레임
def process_docx(path: str):
    doc   = Document(path)
    paras = [p.text.strip() for p in doc.paragraphs if p.text.strip()]

    # 블록 단위 저장용
    blocks = []
    cur_block = None

    for para in paras:
        low = para.lower()
        # 새 블록 시작: Company=
        if low.startswith('company') and '=' in para:
            # 이전 블록이 있으면 저장
            if cur_block:
                blocks.append(cur_block)
            # 새 블록 초기화
            cur_block = {
                'company'     : para.split('=',1)[1].strip().strip('"“”'),
                'year'        : None,
                'article_lines': []
            }
        # 같은 블록의 Year=
        elif low.startswith('year') and '=' in para:
            if cur_block is None:
                raise ValueError(f"{path}: Year 앞에 Company가 없습니다.")
            cur_block['year'] = int(para.split('=',1)[1].strip())
        # 나머지는 Article 텍스트
        else:
            if cur_block:
                cur_block['article_lines'].append(para)

    # 마지막 블록 저장
    if cur_block:
        blocks.append(cur_block)

    # 각 블록별로 감성분석 & a() 호출
    for blk in blocks:
        company = blk['company']
        year    = blk['year']
        lines   = blk['article_lines']

        if company is None or year is None:
            raise ValueError(f"{path}의 블록에 Company/Year 정보가 없습니다: {blk}")

        Article = "\n\n".join(lines)

        # — 기존 문단별 감성분석 로직 재사용 —
        paragraphs = [p for p in Article.split('\n\n') if p.strip()]
        emotion_scores = []
        for para in paragraphs:
            res = sa_kobert(para[:512], truncation=True, padding=True)[0]
            lbl, sc = res['label'], res['score']
            sent = interpret(lbl, sc)
            if   sent == "긍정": emotion_scores.append( sc)
            elif sent == "부정": emotion_scores.append(-sc)
            else:               emotion_scores.append(0)

        non_neutral = [s for s in emotion_scores if s != 0]
        refined_emotion_score = (
            sum(non_neutral)/len(non_neutral) if non_neutral else 0
        )

        # — df_sum/df_count 에 반영 —
        inData(company, year, refined_emotion_score)
        print(f"[{path}] → {company}({year}) = {refined_emotion_score:.4f}")
````

# 여기까지 경영진 평가

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 데이터 불러오기
df = pd.read_csv('merged_df.csv', encoding='utf-8-sig')

# feature 정의
features = [
    'ROE', 'ROA', '영업이익률', '순이익률', 'EPS',    
    '부채비율', '유동비율', '자기자본비율',            
    '매출액 성장률', '순이익 성장률',                 
    '브랜드평판지수',                                
    '경영진평가'                                    
]

# 가중치 정의
feature_weights = {
    'ROE': 3.0, 'ROA': 3.0, '영업이익률': 3.0, '순이익률': 3.0, 'EPS': 3.0,
    '부채비율': 10.0, '유동비율': 10.0, '자기자본비율': 10.0,
    '매출액 성장률': 15.0, '순이익 성장률': 15.0,
    '브랜드평판지수': 15.0,
    '경영진평가': 10.0
}

# 결측치 제거
df_no_nan = df.dropna(subset=features).copy()

# log1p + 가중치 적용
for feature in features:
    log_feature = feature + '_log'
    df_no_nan[log_feature] = np.log1p(df_no_nan[feature].clip(lower=0)) * feature_weights[feature]

# log feature list
log_features = [f + '_log' for f in features]

# StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_no_nan[log_features])

# silhouette score 탐색
scores = []
K_range = range(2, 8)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    scores.append(score)

# 시각화
plt.figure(figsize=(6, 4))
plt.plot(K_range, scores, marker='o')
plt.xlabel('클러스터 수 (k)')
plt.ylabel('Silhouette Score')
plt.title('적절한 클러스터 수 탐색')
plt.grid(True)
plt.show()

# 최적 k 설정 및 클러스터링
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_no_nan['cluster'] = kmeans.fit_predict(X_scaled)

# 클러스터 요약
cluster_summary = df_no_nan.groupby('cluster')[features].mean()
display(cluster_summary)

# PCA 시각화
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df_no_nan['PCA1'] = X_pca[:, 0]
df_no_nan['PCA2'] = X_pca[:, 1]

plt.figure(figsize=(10, 8))
for cluster in sorted(df_no_nan['cluster'].unique()):
    subset = df_no_nan[df_no_nan['cluster'] == cluster]
    plt.scatter(subset['PCA1'], subset['PCA2'], label=f'Cluster {cluster}', alpha=0.6)

plt.title('클러스터링 결과 (PCA)')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend()
plt.grid(True)
plt.show()
````

## 정성적 지표를 포함한 통합 가치투자 클러스터링

본 단계에서는 기존의 재무지표(수익성, 안정성, 성장성)에 더해
브랜드 평판 지수와 경영진 평가 점수라는 정성적 지표를 포함하여
보다 현실적이고 종합적인 투자 판단을 수행할 수 있도록 클러스터링 모델을 확장하였다.

1. 데이터 구성
분석에 사용된 feature는 다음과 같다.

분류	사용된 지표
수익성	ROE, ROA, 영업이익률, 순이익률, EPS
안정성	부채비율, 유동비율, 자기자본비율
성장성	매출액 성장률, 순이익 성장률
정성적 지표	브랜드평판지수, 경영진 평가

결측치를 포함한 행은 제거한 후 클러스터링을 진행하였다.

2. 전처리 및 클러스터링
각 feature에 대해 로그 변환(log1p)을 적용하여 극단값(Outlier)의 영향을 줄였으며,

StandardScaler로 정규화하여 KMeans의 거리 기반 계산이 적절히 작동하도록 하였다.

최적 클러스터 수를 결정하기 위해 2~7개의 클러스터에 대해 Silhouette Score를 비교하고,
그 중 가장 안정적인 구조를 보이는 k=3을 최종 클러스터 수로 선정하였다.

3. 클러스터 해석
각 클러스터의 평균 feature 값을 통해 해당 클러스터의 특성을 정성적으로 분석 가능하다.
예를 들어,

Cluster 0: 브랜드 평판은 높지만 성장성이 낮은 안정적 가치주

Cluster 1: 전반적으로 지표가 낮은 저성장/저수익 기업군

Cluster 2: 브랜드/경영진 지표까지 우수한 고성장 고수익 성장주

cluster_summary를 통해 클러스터별 수치적 특성을 확인할 수 있다.

4. PCA 시각화
고차원 feature를 2차원 공간으로 축소(PCA)하여 각 기업의 클러스터 분포를 시각화하였다.

클러스터별 색상을 다르게 하여 각 기업군의 위치와 유사도를 직관적으로 확인 가능하다.

이 분석은 단순한 재무 기반 평가를 넘어,
시장 신뢰도(브랜드)와 리더십 역량(경영진 평가)까지 반영한
통합 가치투자 모델의 핵심 구조로 활용될 수 있다.

이후 단계에서는 클러스터 해석 결과를 바탕으로 기업별 투자 추천을 도출하게 된다.

````python
# Silhouette Plot 그리기용 라이브러리 import
from sklearn.metrics import silhouette_samples

# Silhouette Plot 그리기 함수 정의
# → 클러스터링 결과의 품질을 시각적으로 평가하는 plot
# → silhouette score가 높을수록 클러스터 내부 응집도는 높고, 클러스터 간 분리도는 좋음
def plot_silhouette(X_scaled, labels, n_clusters):
    # 전체 silhouette 평균 score 계산
    silhouette_avg = silhouette_score(X_scaled, labels)
    
    # 각 샘플의 silhouette coefficient 값 계산
    # → 개별 데이터가 얼마나 잘 해당 클러스터에 속하는지를 나타냄
    sample_silhouette_values = silhouette_samples(X_scaled, labels)

    # 시각화 준비
    import matplotlib.cm as cm
    fig, ax1 = plt.subplots(1, 1)
    fig.set_size_inches(8, 6)

    # y축 위치 초기화 (클러스터별로 쌓아올리기 위해)
    y_lower = 10

    # 클러스터별 silhouette plot 그리기
    for i in range(n_clusters):
        # 현재 클러스터에 속한 데이터의 silhouette 값 추출 및 정렬
        ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        # 클러스터별 색상 설정
        color = cm.nipy_spectral(float(i) / n_clusters)

        # silhouette 값 영역 채우기
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # 클러스터 번호 텍스트 표시
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # 다음 클러스터 영역 위치 설정
        y_lower = y_upper + 10

    # 그래프 제목 및 축 설정
    ax1.set_title(f"실루엣 점수")
    ax1.set_ylabel("클러스터")

    # 전체 silhouette 평균에 수직선 표시
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    # y축, x축 눈금 설정
    ax1.set_yticks([])  # y축 눈금 제거 (클러스터 번호는 텍스트로 표시)
    ax1.set_xticks(np.arange(-0.1, 1.1, 0.2))  # x축 눈금 설정

    # 그래프 출력
    plt.show()

#  실행 예시
# → 현재 최적 k로 학습된 클러스터 결과에 대해 silhouette plot 출력
plot_silhouette(X_scaled, df_no_nan['cluster'], optimal_k)

 silhouette_avg = silhouette_score(X_scaled, labels)
    print(f"Silhouette Score: {silhouette_avg:.3f}")
````

## Silhouette Plot을 활용한 클러스터링 품질 평가

본 분석에서는 최종 선택한 k값으로 KMeans 클러스터링을 수행한 후, **Silhouette Plot**을 활용하여 클러스터링 품질을 시각적으로 평가하였다.

### Silhouette Plot이란?

- **Silhouette Coefficient 값**: 각 데이터 포인트가 얼마나 잘 해당 클러스터에 속해 있는지를 나타내는 지표
    - +1 에 가까울수록 잘 클러스터링됨
    - 0 에 가까울수록 경계에 위치
    - 음수는 잘못 클러스터링됨

- **전체 Silhouette Score 평균값**: 군집 응집도 + 군집 간 분리도를 종합적으로 평가
    - 일반적으로 0.5 이상이면 좋은 품질로 간주
    - 0.25~0.5 정도는 중간 수준
    - 0.25 이하이면 클러스터 품질이 좋지 않음 (재평가 필요)


### 분석 흐름

```plaintext
- 클러스터링 결과(X_scaled, labels)에서 silhouette score 계산
- 각 데이터 포인트의 silhouette coefficient 계산
- 클러스터별로 silhouette value 분포 시각화
- 전체 silhouette 평균값 수직선으로 표시


### <span style="background-color: yellow">분석 결과</span>

###  Silhouette Plot 해석 (k=3 클러스터링 기준)

- **평균 Silhouette Score: 0.2** → 다소 낮은 편 → 클러스터 경계가 완벽하게 명확하진 않음
- **Cluster 2** (고성장 고수익 그룹): 비교적 높은 silhouette 값을 보이며 명확하게 구분된 경향
- **Cluster 1/Cluster 0**: 많은 데이터가 0 근처로 분포 → 클러스터 간 경계가 다소 애매한 경우 존재

#### 종합 결론

- **정성적 지표(브랜드평판, 경영진 평가)** 를 포함한 분석은 의미가 있었으나, 전체 feature의 클러스터 분리도를 더 향상시키기 위한 추가 개선 여지 있음
- 특히 **Cluster 2**는 투자 판단 시 유용한 클러스터로 활용 가능
- **Cluster 0/1**은 feature tuning 또는 feature selection을 통해 **추가적인 분리도 향상 시도 필요**

#### 향후 방향

- 정성적 지표의 **정규화 / 가중치 조정** → Silhouette Score 향상 기대
- feature engineering을 통한 클러스터 간 분리도 개선 필요 (특히 Cluster 0/1 간)




본 분석에서는 KMeans 클러스터링을 적용한 결과, 실루엣 score가 비교적 낮은 값(0.2 이하)을 나타냈다. 이는 분석 데이터의 구조적 특성에 기인한 것으로, 본 데이터는 재무지표(정량적)와 브랜드평판지수 및 경영진 평가(정성적)라는 서로 다른 성격의 feature들이 혼합되어 있다. 또한 워런 버핏 가치투자 기준은 본질적으로 연속적인 가치 스펙트럼을 가지며, 기업 간 명확한 군집 경계가 존재하지 않는다. 이러한 특성으로 인해 KMeans와 같이 중심 기반 클러스터링 알고리즘에서 군집 간 분리가 명확하지 않게 나타났고, 경계선 근처의 데이터 포인트가 많아 실루엣 score가 낮게 형성되었다.

이는 분석 모델의 한계라기보다는 기업 데이터의 본질적 구조를 반영한 결과로 해석할 수 있으며, 오히려 클러스터링을 통해 기업 간 상대적 위치 및 특성을 파악하는 데 중점을 두는 접근이 적절하다고 판단된다.<span style="background-color: yellow">만약 클러스터링 기법을 사용하지 않았다면</span>  개별 feature별 단순 순위 또는 종합점수 기반으로만 기업을 비교할 수 있었을 것이다. 그러나 이러한 접근법은 기업 간 유사성을 파악하거나 시장 내에서 특정 유형의 기업군(예: 안정적 가치주, 고성장 기업군, 저성장·저수익 기업군)의 분포를 이해하는 데 한계가 있다.

특히 정성적 지표(브랜드평판지수, 경영진 평가)는 단순 순위화보다 클러스터링을 통해 통합적 패턴 속에서 해석하는 것이 더 자연스럽고 의미가 있다. 따라서 클러스터링 기법은 다양한 재무적·정성적 지표를 종합적으로 고려한 기업 특성 유형화를 가능하게 하며, 이를 통해 보다 깊이 있는 투자 인사이트를 도출하는 데 기여한다고 볼 수 있다.


````python
# 라이브러리 import
import matplotlib.pyplot as plt
import numpy as np

# 상위 N개 기업 설정
# → 브랜드평판지수 기준으로 상위 N개 기업만 표시
N = 100

# 브랜드평판지수 기준 상위 N개 기업 추출
# → 브랜드 가치가 높은 기업부터 N개 추출 (정렬 후 head 사용)
top_companies = df_no_nan.sort_values('브랜드평판지수', ascending=False).head(N)

# PCA scatter plot 그리기
plt.figure(figsize=(12, 10))  # 그림 크기 설정

# 클러스터별 scatter plot
# → 각 클러스터별로 다른 색상으로 표시
for cluster in sorted(df_no_nan['cluster'].unique()):
    subset = df_no_nan[df_no_nan['cluster'] == cluster]
    plt.scatter(subset['PCA1'], subset['PCA2'], label=f'Cluster {cluster}', alpha=0.6)

# 기업명 + 분기라벨 표시 (Top N 기업 중복 방지)
# → 표시 시 텍스트가 겹치는 것을 방지하기 위해 offset 적용
# → 동일 기업명은 최초 1회만 표시

# seed 고정 → 동일 실행 시 동일 결과 재현 가능
np.random.seed(42)

# 표시한 기업명 기록용 set 생성
displayed_companies = set()

for i, row in top_companies.iterrows():
    company_name = row['기업명']
    
    # 이미 표시한 기업명은 skip
    if company_name in displayed_companies:
        continue
    
    # 텍스트 구성: 기업명 + (분기라벨)
    label_text = f"{company_name} ({row['분기라벨']})"
    
    # offset 설정 → 텍스트 겹침 방지 (약간의 random 위치 조정)
    x_offset = np.random.uniform(-0.1, 0.1)
    y_offset = np.random.uniform(-0.1, 0.1)
    
    plt.text(row['PCA1'] + x_offset, row['PCA2'] + y_offset, label_text, fontsize=8, alpha=0.9)
    
    # 표시한 기업명 set에 추가
    displayed_companies.add(company_name)

# plot 마무리 설정
plt.title(f'클러스터링 결과')
plt.xlabel('PCA1')   # X축 레이블
plt.ylabel('PCA2')   # Y축 레이블
plt.legend()         # 범례 표시 (클러스터별 색상 구분)
plt.grid(True)       # grid 표시
plt.show()           # plot 출력
````

이를 통해 브랜드 경쟁력이 높은 기업들이 정량적 특성 기반으로 어떠한 클러스터에 위치하는지, 그리고 이러한 기업들이 버핏식 가치투자에서 추구하는 방어적 가치주 또는 고성장 기대주와 어떤 관계를 가지는지를 직관적으로 확인할 수 있다.

````python
# 필요 라이브러리 import
from IPython.display import display, HTML

# 출력할 컬럼 지정
columns_to_display = ['기업명', '브랜드평판지수', 'cluster']

# Cluster 특성 설명 사전 정의
cluster_description = {
    0: "안정적 가치주 (버핏형 가치주, 장기 투자 적합)",
    2: "고성장·고수익 기업군 (성장주, 리스크 있음, 신중 투자)",
    1: "저성장·저수익 기업군 (투자 비추천 또는 신중 접근)"
}

# 클러스터 우선 순서 정의
cluster_order = [0, 2, 1]

# 이미 출력한 기업명 기록용 set
printed_companies = set()

# HTML 코드 모음용 리스트
html_tables = []

for cluster_id in cluster_order:
    # 해당 클러스터 데이터만 추출
    cluster_data = df_no_nan[df_no_nan['cluster'] == cluster_id]
    
    # 브랜드평판지수 기준 내림차순 정렬
    cluster_data_sorted = cluster_data.sort_values('브랜드평판지수', ascending=False)
    
    # 중복 제거 + 우선순위 적용 → 아직 출력되지 않은 기업만 남기기
    cluster_unique = cluster_data_sorted[~cluster_data_sorted['기업명'].isin(printed_companies)]
    
    # 기업명 기준 중복 제거 후 상위 20개 추출
    cluster_top_20 = (
        cluster_unique
        .drop_duplicates(subset='기업명', keep='first')
        .head(20)
    )
    
    # 출력한 기업명 set에 추가
    printed_companies.update(cluster_top_20['기업명'].tolist())
    
    # DataFrame을 HTML로 변환
    table_html = cluster_top_20[columns_to_display].reset_index(drop=True).to_html(index=False)
    
    # 제목과 특성 설명 포함 → HTML로 구성
    full_html = f"""
    <div style="display: inline-block; vertical-align: top; margin-right: 20px;">
        <h3>Cluster {cluster_id}</h3>
        <p><b>특성:</b> {cluster_description.get(cluster_id, '설명 없음')}</p>
        {table_html}
    </div>
    """
    
    # 리스트에 추가
    html_tables.append(full_html)


display(HTML(''.join(html_tables)))
````
